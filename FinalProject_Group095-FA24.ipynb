# Abstract

This study examines the relationship between public sentiment on COVID-19 and vaccination rates, focusing on the most populous countries. Public health measures such as mask mandates, lockdowns, and vaccination requirements generated polarized views during the pandemic, and these sentiments could be linked to vaccination rates differently across regions. Here, we analyze tweet and vaccine data from China, India, and the United States, and explore if and how twitter sentiment is correlated to vaccination rates in these countries, and furthermore how these correlations differ. Through the utilization of VADER sentiment analysis, Pearson correlation calculation, and Fisher's z-transformation statistical analysis, we found that tweet sentiment and vaccination rates were significantly correlated in India and China, but not in the United States. Additionally, we found that these correlations were significantly different between India and China, as well as between India and the United States.

Does Twitter sentiment regarding COVID-19 public health measures correlate with vaccination rates differently in the three most populous countries (the United States, India, and China) from December 2020 to November 2021?

Is there a significant correlation between positive sentiments expressed on Twitter regarding COVID-19 public health measures and the vaccination rates during the COVID-19 pandemic? Conversely, do negative sentiments correlate with lower vaccination rates? How do these relationships differ between these three countries?

## Background and Prior Work


Public health measures introduced in response to the COVID-19 pandemic, such as mask mandates, lockdowns, and vaccination requirements, have not only shaped government policy but have also spurred considerable public debate. Measures that aimed to control viral spread sparked polarized views that were shaped by multiple external factors including political affiliation and region-specific COVID-19 cases. As the pandemic evolved, so too did their policies. The ongoing effect that these policies have continues to impact public sentiment. 

Analyzing such public sentiments provides insight into broader societal perspectives on COVID-19-specific public health policies and offers valuable information for shaping future public health responses. X (formerly Twitter) has been central forum<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) where the public has discussed COVID-19 health policies during the pandemic. They are accessible online spaces where users can create, modify, and express varying sentiments about such health measures. Prior studies demonstrate that public sentiment on these platforms fluctuates in response to major world events or policy changes. Such public sentiments vary due to factors such as region or platform. For instance, one study analyzed public sentiments on Twitter<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2), posted within 14 days between January 14 – 28, 2020, and selected based on whether they contained keywords from a generated list of nineteen hashtags from Brazil, India, Mexico, Oman, Peru, Russia, Spain, the United Kingdom, and the United States. This study concluded that countries generally had different sentiments due to their geographic boundaries: for instance, in the United States, keywords such as “‘trump’ [sic], ‘Pandemic’ and ‘Death’ [were] associated with ‘Surprise,’ ‘Sadness,’ and ‘Disgust’”; in Peru, keywords such as “‘pandemic,’ ‘stay home’ and ‘distance’ [were associated] with the emotions of sadness and joy.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)”  This study demonstrates that public sentiment regarding COVID-19 public health policies differs and evolves based on region, changing health conditions and public policies, and whether they were expressed during the pandemic or post-pandemic periods. 

Our study builds on this prior research by analyzing the following: is there a correlation between positive public sentiment on COVID-19 public health policies and higher vaccination rates within a region, and conversely, is there a correlation between negative public sentiment on COVID-19 public health policies and lower vaccination rates? Our project aims to offer insights on how public sentiment trends in specific regions correlate with vaccination rates, as these trends could potentially inform future public health policy and communication strategies.

## Descriptions of Datasets and Combining Data
### Dataset #1: COVID-19 All Vaccines Tweets
This dataset contains a collection of recent tweets about COVID-19 vaccines used internationally on a large scale. These tweets reference vaccinations such as Pfizer/BioNTech, Sinopharm, Sinovac, Moderna, Oxford/AstraZeneca, Covaxin, and Sputnik V. This dataset contains several variables; the variables listed below are the ones relevant to this study: 
- **'date':** presented as a datetime datatype as a metric of the date on which the tweet was posted. This is a proxy for time series data, which allows for temporal analysis of public sentiment and trends related to the aformentioned COVID-19 vaccines. 
- **'user_location':** presented as a string datatype using the location of the user measuring the geographic origin of the tweet (varying in specificity from city to country). This is a proxy for geographic origin of sentiment and regional attitudes towards COVID-19 policies, vaccination rates, and health concerns. 
- **'text':** presented as a string datatype with the contents of the tweets. This will be the main focus of our investigation with sentiment analysis. This is a proxy for public sentiment towards COVID-19 vaccines with more specificity than just hashtags. The resulting sentiment scores derived from tokenizing and analyzing these tweets would act as proxies for emotional responses towards the topic in question.
- **'hashtags':** presented as a string datatype using a select list of hashtags that can be quantified to measure the topics emphasized in the tweet (for example, "covid19" or "coronavirus"). Hashtags will not be analyzed directly themselves but will be used as an extra measure of filtering to ensure the tweets being analyzed are relevant to our topic, and that analysis is not being confounded by unrelated public opinions. By analyzing the text corresponding with the relevant hashtag, sentiment analysis can be performed to analyze opinions on vaccines, COVID-19, and associated policies. 

To wrangle this dataset, functions were made to clean the data to only show the variables relevant to the study. These included the date from which the tweets were posted and relevant city and country location from where the tweets were posted (China, India, and the United States). For hashtag data, this data column was cleaned by removing URLs, mentions, hashtags, and stripping excess whitespace before tokenization. Before the data is processed by sentiment analysis techniques, we will tokenize it to visualize word frequency and confirm relevance of tweet texts.

### Dataset #2: COVID-19 World Vaccination Progress
This dataset from Our World in Data's COVID-19 data repository contains time-series data specifically on COVID-19 vaccination efforts conducted across various countries around the world. We have wrangled this dataset down to only vaccination data from China, India and U.S.A. to suit the purposes of our study. The following variables are the ones relevant to this study:
- **'country':** presented as a string datatype that marks the country that the vaccination data is from (amongst China, India or U.S.A.).
- **'date':** presented as a datetime datatype to measure the timeline for vaccination progress. This is a proxy for the timeline of vaccination progress in the country of interest, enabling studiers to track vaccination rate changes in correlation with public health events (e.g., the Omicron variant) or changes in public health policies.
- **'daily_vaccinations':** presented as an integer datatype measuring the daily count of vaccinations administered, including secondary shots and boosters. It is **not** the number of patients that have been vaccinated. This is an important distinction, as total vaccination counts *will* exceed country populations, and this is reasonable given the vaccination protocols. This variable can nonetheless be a proxy for the partial immunity within the population and reflects the progress of the country's vaccination rates. These metrics can be compared and potentially correlated with reductions in COVID-19 case rates.

```python
# Functions to clean user location and match location to countries/states

def clean_user_location(user_location):
    if pd.isna(user_location): 
        return None
    cleaned_location = re.sub(r'[^a-zA-Z\s]', '', user_location).strip().lower()
    location_parts = cleaned_location.split()

    return location_parts

def match_location(user_location):
    if not user_location:
        return None
    
    matched_location = None
    for location in user_location:  
        for country, locations in location_dict.items():
            if location in locations: 
                matched_location = country
                break
        if matched_location:
            break

    return matched_location if matched_location else None

# Applying above functions to dataframe
tweets = og_tweets[['date', 'user_location', 'text', 'hashtags']].copy()
tweets['user_location_elements'] = tweets['user_location'].apply(clean_user_location)
tweets['country'] = tweets['user_location_elements'].apply(match_location)

tweets = tweets.dropna(subset=['country'])

# Date formatting, dropping timestamps
tweets['date'] = pd.to_datetime(tweets['date'], errors='coerce').dt.date

# Creating final dataframe for further text wrangling
tweets = tweets[['date', 'country', 'text', 'hashtags']]

# Checking null and missing values
# Expecting no missing values for country, possible missing for hashtags
print("\nMissing values count for key columns:")
print(tweets[['date', 'country', 'text', 'hashtags']].isnull().sum())

tweets = tweets.reset_index(drop=True)

print(tweets.shape)
tweets.head()
```

```python
# Cleaning texts before tokenization

def extract_hashtags(text):
    hashtags = re.findall(r'#\w+', str(text))
    if not hashtags:
        return np.nan
    return hashtags

tweets['hashtag_from_tweet'] = tweets['text'].apply(extract_hashtags)

def prepare_tweet(tweet):
    """
    Lowercases text
    Removes URLs 
    Removes mentions
    Removes hashtags 
    Strips excess whitespace
    (before tokenization)

    """
    try:
        tweet = tweet.lower()
        tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)
        tweet = re.sub(r'@\w+', '', tweet)
        tweet = re.sub(r'#\w+', '', tweet)
        tweet = tweet.strip()

    except:
        tweet = np.nan

    return tweet

tweets['text'] = tweets['text'].apply(prepare_tweet)
# tweets['hashtags'].fillna(tweets['hashtag_from_tweet'], inplace=True)
# tweets['hashtags'] = tweets['hashtags'].replace('', np.nan)


# Filtering data for relevance by hashtags
# First all hashtags are extracted and converted to an appropriate datatype for further parsing
# Then only valid hashtags are kept (if they are in the valid_keywords list)
# NaN observations for hashtags are dropped if neither hashtag nor tweet contains valid keywords
# Ensures data is very specific to our topic

valid_keywords = ['corona', 'coronavirus', 'covid', 'covid19', 'vaccine', 'covidvaccine','covid19vaccine', 
                  
                  # Specific vaccines
                  'biontech', 'moderna', 'astrazeneca', 'sinovac','sinopharm', 'sputnikv', 'covaxin']

def clean_hashtags(hashtags):
    if isinstance(hashtags, list): 
        cleaned_hashtags = []
        for hashtag_str in hashtags:
            if isinstance(hashtag_str, str):  # Ensure each item is a string
                hashtag_str = hashtag_str.lower()
                hashtags_list = hashtag_str.split()
                cleaned_hashtags.extend([hashtag.lstrip('#') for hashtag in hashtags_list if hashtag.lstrip('#') in valid_keywords])
        return cleaned_hashtags if cleaned_hashtags else np.nan
    return np.nan 

def contains_valid_keywords(tweet):
    tweet = str(tweet).lower()
    return any(keyword in tweet for keyword in valid_keywords)

# Applying the clean_hashtags function
tweets['hashtag_from_tweet_checked'] = tweets['hashtag_from_tweet'].apply(clean_hashtags)

tweets = tweets[~(tweets['hashtag_from_tweet_checked'].isna() & tweets['text'].apply(lambda x: not contains_valid_keywords(x)))]

# Cleaning up the dataframe
tweets.drop('hashtags', axis=1, inplace=True)
tweets.drop('hashtag_from_tweet', axis=1, inplace=True)
tweets = tweets.rename(columns={'hashtag_from_tweet_checked': 'filtered_hashtags'})

print("Null counts for each variable") # Hashtags being null is okay, they have been filtered
print(tweets.isnull().sum())
print()

print("Shape of dataframe after first round of cleaning")
print(tweets.shape)
print()

print("Tweet counts per country") # Note that China has significantly less tweets than US or India, this will be in our analysis
print(tweets['country'].value_counts()) 

print()
tweets.head()
```

### Section 1: VADER Sentiment Analysis on Tweet Data

In this section, we are chunking Twitter text data to feed into VADER. We aim to compute sentiment scores for tweets per country per week. Then we will be visualizing these scores over time for each of the nations to get our first understanding of the differences in sentiment across our countries of interest. We will keep in mind that the amount of Twitter data we have for China is substantially less than that for the United States and India. 

```python
# Chunking Twitter data for VADER into per-week categories

# Finding the earliest and latest tweet date per country
date_range_per_country = (
    tweets.groupby('country')['date']
    .agg(earliest = 'min', latest = 'max')
    .reset_index()
)

print(date_range_per_country)
```

```python
tweets['date'] = pd.to_datetime(tweets['date'])

# Uniform earliest and latest dates
earliest_date = pd.Timestamp("2020-12-12")
latest_date = pd.Timestamp("2021-11-23")

# Generating weekly ranges
weeks = []
current_start = earliest_date
week_number = 1

while current_start <= latest_date:
    current_end = current_start + timedelta(days = 6)
    weeks.append((f"Week {week_number}: {current_start.date()} - {current_end.date()}", current_start, current_end))
    current_start = current_end + timedelta(days = 1)
    week_number += 1

# Creating dictionaries for each country
china_tweets, india_tweets, us_tweets = {}, {}, {}

# Grouping tweets by country
for country, country_name in [('china', china_tweets), ('india', india_tweets), ('us', us_tweets)]:
    for week_label, start_date, end_date in weeks:

        filtered_tweets = tweets[
            (tweets['country'] == country) &
            (tweets['date'] >= pd.Timestamp(start_date)) &
            (tweets['date'] <= pd.Timestamp(end_date))
        ]
        
        # Assigning the tweets to the corresponding week key
        country_name[week_label] = filtered_tweets['text'].tolist()

# # Checking first week data for each country
# print("China Tweets:", list(china_tweets.items())[:1])
# print("India Tweets:", list(india_tweets.items())[:1])
# print("US Tweets:", list(us_tweets.items())[:1])
```

```python
#Initializing the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Computing  average sentiment for a week
def compute_weekly_sentiment(tweet_list):
    if not tweet_list:
        return None
    scores = [sia.polarity_scores(tweet)['compound'] for tweet in tweet_list]
    return sum(scores) / len(scores)

# Analyzing sentiment for each country
china_sentiment = {week: compute_weekly_sentiment(tweets) for week, tweets in china_tweets.items()}
india_sentiment = {week: compute_weekly_sentiment(tweets) for week, tweets in india_tweets.items()}
us_sentiment = {week: compute_weekly_sentiment(tweets) for week, tweets in us_tweets.items()}

# Displaying example output
print("China Sentiment:")
for week, sentiment in list(china_sentiment.items())[:5]:
    print(f"{week}: {sentiment}")

print("\nIndia Sentiment:")
for week, sentiment in list(india_sentiment.items())[:5]:
    print(f"{week}: {sentiment}")

print("\nUS Sentiment:")
for week, sentiment in list(us_sentiment.items())[:5]:
    print(f"{week}: {sentiment}")
```

### Section 3: Vaccine Rate and Twitter Sentiment Correlations

In this section, we use the `pearsonr` correlation coefficients and p-values to assess whether weekly Twitter sentiment is positively, negatively, or not correlated with weekly vaccination rate. We will then use **Ronald Fisher's z-transformation** to normalize the correlation coefficients in order to directly compare and determine whether correlations are significantly different between countries.

```python
# Import necessary packages
import scipy as sp
from scipy.stats import pearsonr
from scipy.stats import norm
```

```python
# Computing the correlation coefficient (r) and p-value for the relationship between Twitter sentiment and vaccination rate in India.
sentiments_array_india = np.array([float(x) if x is not None else 0.0 for x in list(india_sentiment.values())])
rate_array_india = np.array(india_df_eda['avg_daily_rate'], dtype=float)

r_india, p_india = sp.stats.pearsonr(sentiments_array_india, rate_array_india)

print(f'The correlation between Twitter sentiment and vaccination rate in India is {r_india} with a p-value of {p_india}.')
print(f'This {'is' if p_india <= 0.05 else 'is not'} significant at 95% confidence (alpha = 0.05).')
print(f'This {'is' if p_india <= 0.01 else 'is not'} significant at 99% confidence (alpha = 0.01).')
```

```python
# Computing the correlation coefficient (r) and p-value for the relationship between Twitter sentiment and vaccination rate in China.
sentiments_array_china = np.array([float(x) if x is not None else 0.0 for x in list(china_sentiment.values())])
rate_array_china = np.array(china_df_eda['avg_daily_rate'], dtype=float)

r_china, p_china = sp.stats.pearsonr(sentiments_array_china, rate_array_china)

print(f'The correlation between Twitter sentiment and vaccination rate in China is {r_china} with a p-value of {p_china}.')
print(f'This {'is' if p_china <= 0.05 else 'is not'} significant at 95% confidence (alpha = 0.05).')
print(f'This {'is' if p_china <= 0.01 else 'is not'} significant at 99% confidence (alpha = 0.01).')
```

```python
# Defining a function to normalize the correlation coefficient.
def fisher_z(r):
    return 0.5 * np.log((1 + r) / (1 - r))

# Defining a function to compare two normalized correlation coefficients.
def compare_correlations(r1, n1, r2, n2):
    z1 = fisher_z(r1)
    z2 = fisher_z(r2)

    se = np.sqrt(1 / (n1 - 3) + 1 / (n2 - 3))
    z_stat = (z1 - z2) / se
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    return z_stat, p_value
```

```python
# Comparing the correlation coefficients for China and India.
z_india_china, p_india_china = compare_correlations(r_china, len(sentiments_array_china), r_india, len(sentiments_array_india))

print(f'The z-score difference between the Twitter sentiment/vaccination rate correlations of India and China is {z_india_china} with a p-value of {p_india_china}.')
print(f'This {'is' if p_india_china <= 0.05 else 'is not'} significant at 95% confidence (alpha = 0.05).')
print(f'This {'is' if p_india_china <= 0.01 else 'is not'} significant at 99% confidence (alpha = 0.01).')
```

We as a collective recognize that tweets during the COVID-19 pandemic often reflect deeply personal or traumatic experiences. While our sentiment analysis focuses on identifying trends, it does not aim to diminish or misinterpret these lived experiences. Our findings are strictly observational and do not seek to infer causality. We approach this work with empathy and respect for the individuals whose voices are represented in the data.

Tweets are shared on a public platform, but ethical research demands that we prioritize user privacy. To uphold this, all datasets will be anonymized before analysis. Identifiable information, such as usernames or profile details, will be excluded. Our focus will remain on non-identifiable attributes, such as region, tweet text, post metrics, and date of posting. This approach aligns with ethical research practices and complies with data protection regulations.

This approach fosters diverse perspectives and reduces the influence of individual biases.

Our research adheres to ethical standards and legal requirements for data usage. All datasets are sourced in accordance with Twitter's terms of service and relevant data protection laws. # Discussion and Conclusion 

This project examined the relationship between public sentiment about COVID-19 public health measures on Twitter and vaccination rates across three of the world’s most populous countries — India, China, and the United States — during the critical period from December 2020 to November 2021. Utilizing tools such as VADER sentiment analysis, Pearson correlation, and exploratory data visualization, we aimed to uncover whether public sentiment trends were significantly associated with vaccination progress and how these associations differed across countries.

## Key Findings

Our results revealed some notable differences in both public sentiment and vaccination trends between the three countries:. 
- **United States:** Twitter sentiment in the U.S. displayed a predominantly positive trend throughout the study period, with no recorded weeks of negative sentiment. Vaccination rates peaked in mid-April 2021, the earliest peak among the three countries. Figure 3 shows that this coincided with a spike in tweet activity in April 2021, followed by a sharp decline, potentially reflecting a shift in public discourse away from vaccination topics after widespread initial vaccinations. When both Twitter and vaccination analyses were consolidated and analyzed using Pearson correlation, we find that the relationship between Twitter sentiment over time and vaccination rate over time is given by a weak r = -0.081 and is not statistically significant at 95% confidence nor 99% confidence. These results suggest that COVID-19 vaccination rates in the U.S. during the studied period were not strongly influenced by public sentiment on Twitter. Consequently, we fail to reject the null hypothesis, which posits a lack of correlation between sentiment and vaccination rates in the U.S.
- **India:** Twitter sentiment exhibited higher variability in India, with a brief period of starkly negative sentiment recorded in January of 2021. However, by Figure 2, India’s Twitter attitudes towards the vaccination became comparatively more positive than where it started and more so than the US or China’s by the end of November 2021. While cumulative vaccination counts rose steadily in India over the given period, vaccination rate peaked just before October 2021. Notably, this period did not correspond to a significant change in tweet counts but did coincide with a slight positive shift in Twitter sentiment. Our Pearson correlation analysis revealed a positive correlation at r = 0.664 between Twitter sentiment and vaccination rates in India, significant at both 95% and 99% confidence levels. These outcomes suggest that Twitter sentiment over December 2020 to November 2021 can be associated with the COVID-19 vaccination rates during this period. Thus, we reject the null hypothesis for India, and accept the alternative hypothesis that vaccination rates are correlated with Twitter sentiment in India.
- **China:** The analysis of China was constrained by limited data availability, resulting in less definitive conclusions. In Figure 2, it is immediately noticeable that China faces a sharp decline to negative sentiment in January of 2021, but remains tightly fluctuating between a positive and neutral sentiment thereafter. Trends in tweet counts were less discernible due to sparse data points. Vaccination rates in China peaked during June and July 2021; however, no clear corresponding shifts in sentiment could be visualized. Inferential analysis provided more insight, showing a moderate positive correlation at r = 0.307 between Twitter sentiment and vaccination rates, statistically significant at the 95% confidence level but not at 99%. This indicates that while there was some association between sentiment and vaccination rates in China, it was less robust compared to India. Therefore, we reject the null hypothesis at 95% confidence but fail to do so at 99% confidence.

#### Differences in Tweet-Vaccine Rate Correlations between Countries: 
The analysis of statistically significant differences between countries using Fisher’s z-transformation revealed notable contrasts in the strength of correlations. The difference in the strength of the correlation of Twitter sentiment and vaccination rates was statistically significant between India and China, robustly at 95% confidence, but not at 99% confidence. Similarly, India demonstrated a significantly stronger correlation compared to the United States, achieving significance at both 95% and 99% confidence thresholds. By contrast, the difference in correlations observed between China and the United States were not statistically significant at either threshold, indicating comparably weak or moderate associations between Twitter sentiment and vaccination rates in these two countries. These findings emphasize the unique relationship between public sentiment and vaccination rates across the nations of interest. However, we do acknowledge that such results may be a consequence of varying data availability between the three countries.

#### Hypothesis Testing
Our project highlights that the relationship between public sentiment and vaccination rates varied significantly across countries, reflecting different socio political and cultural contexts. In India, the strong correlation suggests that public sentiment on Twitter played a significant role in driving vaccination rates, potentially influenced by governmental campaigns and public discourse. In contrast, the weaker or non-significant correlations in the U.S. and China indicate that other factors—such as pre-existing public health infrastructure, vaccine availability, or governmental policies—may have had a more dominant influence. Overall, these findings underscore the importance of tailoring public health messaging and interventions to regional contexts to maximize vaccination uptake.


### Limitations 

While our analysis suggests some correlation between Twitter sentiment and vaccination rates in India and China and lesser so in the U.S., there are many limitations in our methodology we must acknowledge. First, the data from China was limited compared to India and the US, making it hard to draw broad conclusions or extract definitive results. Twitter is not as a popular nor accessible type of public forum in China, and the subsection of the population that uses it in English is even finer. Thus, we acknowledge that our results for China may not be entirely representative of its full demographic. In this vein, Twitter is not the *main* form of social media people consume and utilize in each country and across all age groups, so the voices we captured may not fully represent each country’s full population. We also relied heavily on English language sentiment tools such as VADER, which may miss cultural nuances or language-specific expressions that were not included in our analysis.

Beyond these data and linguistic challenges, our study is correlational, not causal. We can’t say for certain that positive tweets led to more vaccinations. Furthermore, because we grouped data by weeks, we might have overlooked short-term spikes or events that happened within those weeks that led to increases and decreases of vaccination rates or sentiment scores. And of course, policy changes, media coverage, and other external factors likely influenced both public sentiment and vaccination rates. Because of different COVID-19 restrictions and vaccination policies in different countries, we could not have controlled for vaccination procedures and requirements. So, while one nation may have mandated only one dose or released the vaccine to the public at an earlier time or had different barriers to administration, the same may not be said for the other nations. Lastly, we cannot ascertain that a third variable —— such as differences in population density, number of active COVID cases, deaths in the country —— was not the reason for the trends we see above. Nonetheless, this exploration opened up avenues for understanding how public health policies and individual action is affected by public sentiment. 


### Conclusion

This project contributes to understanding the complex interplay between public sentiment and health behaviors during a global crisis. While our findings indicate that Twitter sentiment may correlate with vaccination trends, they also highlight the need for more inclusive analyses. Looking ahead, we hope that future researchers can address these issues by incorporating data from other platforms and official sources, refining sentiment analysis tools to order to handle multiple languages and cultural contexts, and employing methods that help us better identify what truly drives changes in vaccination trends. By building on this work, we hope that researchers can develop targeted strategies to improve vaccine uptake and public health outcomes in diverse contexts.
